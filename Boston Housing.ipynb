```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('ggplot')
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
bsdata = load_boston()
df = pd.DataFrame(bsdata.data, columns=bsdata.feature_names)
df['MEDV'] = bsdata.target #データを見てみる
df
```

```python
X = df.iloc[:, 0:13].values
y = df['MEDV'].values
df
```

```python
x=df.RM.to_frame()
y=df.MEDV
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split

(X_train, X_test, y_train, y_test) = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#ここからは重回帰

# Std scalerを使って標準化する
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
X_train_std[0]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X_train_std, y_train)
```

```python
print('傾き', model.coef_)
print('切片', model.intercept_)
```

```python
y_train_pred = model.predict(X_train_std)
y_test_pred = model.predict(X_test_std)
```

```python
print('MSE train: %.2f, test: %.2f' % (np.mean((y_train - y_train_pred) ** 2), np.mean((y_test - y_test_pred) ** 2)))
```

```python
plt.figure(figsize=(8, 4))

plt.scatter(y_train_pred, y_train_pred - y_train, c='red', marker='o', edgecolor='white', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='blue', marker='s', edgecolor='white', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, color='black', lw=2)
plt.xlim([-10, 50])
plt.tight_layout()

plt.show()
```

```python
# LightGBM
import lightgbm as lgb
 
# Scikit-learn（評価算出）
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
hyper_params = {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': ['l2', 'auc'],
    'learning_rate': 0.005,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.7,
    'bagging_freq': 10,
    'verbose': 0,
    "max_depth": 8,
    "num_leaves": 128,  
    "max_bin": 512,
    "num_iterations": 100000,
    "n_estimators": 1000
}
```

```python
gbm = lgb.LGBMRegressor(**hyper_params)
```

```python
gbm.fit(X_train, y_train,
        eval_set=[(X_test, y_test)],
        eval_metric='l1',
        early_stopping_rounds=1000)
```

```python
y_pred_gbm = gbm.predict(X_train, num_iteration=gbm.best_iteration_)
```

```python
# Basic RMSE
from sklearn.metrics import mean_squared_log_error
print('The rmse of prediction is:', round(mean_squared_log_error(y_pred_gbm, y_train) ** 0.5, 5))
```
